```python
import pandas as pd
df = pd.read_excel('Employee.xlsx',index_col='Name')
df
```

## ใส่ช่องว่างด้วยค่าเฉลี่ย

```python
df.describe() # หาค่าเฉลี่ย
select = 'Salary' # เลือกคอลัม
df
```

```python
# fillna() นำข้อมูลไปแทนที่ , mean() นำค่าเฉลี่ยมาใส่ 
df[select] = df[select].fillna(df[select].mean())
df
```

## ใส่ช่องว่างด้วยค่ากำหนดเอง

```python
df = pd.read_excel('Employee.xlsx',index_col='Name')
df
```

```python
select = 'Salary'
df[select] = df[select].fillna(180)
df
```
