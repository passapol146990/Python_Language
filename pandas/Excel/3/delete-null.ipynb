## ลองลบ ตำแหน่งกับเงินเดือนที่ว่างออก --> กำหนดค่าว่างนำค่าก่อนหน้ามาใส่ --> กำหนด Address ถ้าว่างให้ใส่ KLS

```python
df = pd.read_excel('Employee.xlsx',index_col='Name')
df
```

```python
df.dropna(subset=['Job','Salary'],inplace=True)
df
```

```python
df.fillna(method='pad',inplace=True)
df['Address'] = df['Address'].fillna('KLS')
df
```
